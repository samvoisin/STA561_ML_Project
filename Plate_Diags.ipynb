#### Artificial Neural Network (ANN) Plate Diagrams

```python
import numpy as np
import pandas as pd
import daft
import matplotlib.pyplot as plt
```

```python
graphOrigin = (0, -0.5)
graphShape = (8, 6)
inpnodeSpace = 0.55 # margin between input nodes
hidnodeSpace = 0.65 # margin between hidden nodes
outnodeSpace = 0.75 # margin between input nodes
layerSpace = 1.4
```

### Single hidden layer from paper

```python
pgm = daft.PGM(shape = graphShape, origin = graphOrigin)

# input nodes; these are the channels
inplayerStartX = 0.5
inplayerStartY = 0
pgm.add_node(daft.Node("C1",
                       r"$C_{1}$",
                       inplayerStartX,
                       inplayerStartY + 1 * inpnodeSpace,
                       observed = True))
pgm.add_node(daft.Node("C2",
                       r"$C_{2}$",
                       inplayerStartX,
                       inplayerStartY + 2 * inpnodeSpace,
                       observed = True))
pgm.add_node(daft.Node("C3",
                       r"$C_{3}$",
                       inplayerStartX,
                       inplayerStartY + 3 * inpnodeSpace,
                       observed = True))
pgm.add_node(daft.Node("C4",
                       r"$C_{4}$",
                       inplayerStartX,
                       inplayerStartY + 4 * inpnodeSpace,
                       observed = True))
pgm.add_node(daft.Node("C5",
                       r"$C_{5}$",
                       inplayerStartX,
                       inplayerStartY + 5 * inpnodeSpace,
                       observed = True))
pgm.add_node(daft.Node("C6",
                       r"$C_{6}$",
                       inplayerStartX,
                       inplayerStartY + 6 * inpnodeSpace,
                       observed = True))
pgm.add_node(daft.Node("C7",
                       r"$C_{7}$",
                       inplayerStartX,
                       inplayerStartY + 7 * inpnodeSpace,
                       observed = True))
pgm.add_node(daft.Node("C8",
                       r"$C_{8}$",
                       inplayerStartX,
                       inplayerStartY + 8 * inpnodeSpace,
                       observed = True))

# hidden layer1
hidlayerStartX = inplayerStartX + layerSpace
hidlayerStartY = 0.1
pgm.add_node(daft.Node("H1",
                       r"$H_{1}$",
                       hidlayerStartX,
                       hidlayerStartY + 0 * hidnodeSpace))
pgm.add_node(daft.Node("H2",
                       r"$H_{2}$",
                       hidlayerStartX,
                       hidlayerStartY + 1 * hidnodeSpace))
pgm.add_node(daft.Node("H3",
                       r"$H_{3}$",
                       hidlayerStartX,
                       hidlayerStartY + 2 * hidnodeSpace))
pgm.add_node(daft.Node("H4",
                       r"$H_{4}$",
                       hidlayerStartX,
                       hidlayerStartY + 3 * hidnodeSpace))
pgm.add_node(daft.Node("H5",
                       r"$H_{5}$",
                       hidlayerStartX,
                       hidlayerStartY + 4 * hidnodeSpace))
pgm.add_node(daft.Node("H6",
                       r"$H_{6}$",
                       hidlayerStartX,
                       hidlayerStartY + 5 * hidnodeSpace))
pgm.add_node(daft.Node("H7",
                       r"$H_{7}$",
                       hidlayerStartX,
                       hidlayerStartY + 6 * hidnodeSpace))
pgm.add_node(daft.Node("H8",
                       r"$H_{8}$",
                       hidlayerStartX,
                       hidlayerStartY + 7 * hidnodeSpace))

# Output layer
## NOTE: The paper includes 2 fixed bias components here to accout for complex gestures.
## We should consider what this means for us
outlayerStartX = hidlayerStartX + layerSpace
outlayerStartY = 1.2
pgm.add_node(daft.Node("O1",
                       r"$O_{1}$",
                       outlayerStartX,
                       outlayerStartY + 0 * outnodeSpace))
pgm.add_node(daft.Node("O2",
                       r"$O_{2}$",
                       outlayerStartX,
                       outlayerStartY + 1 * outnodeSpace))
pgm.add_node(daft.Node("O3",
                       r"$O_{3}$",
                       outlayerStartX,
                       outlayerStartY + 2 * outnodeSpace))
pgm.add_node(daft.Node("O4",
                       r"$O_{4}$",
                       outlayerStartX,
                       outlayerStartY + 3 * outnodeSpace))

# activation function
actlayerStartX = outlayerStartX + layerSpace
actlayerStartY = ((outlayerStartY + 1 * outnodeSpace) + (outlayerStartY + 2 * outnodeSpace)) / 2
pgm.add_node(daft.Node("Act",
                       r"$A$",
                       actlayerStartX,
                       actlayerStartY + 0 * outnodeSpace))

# prediction
predX = actlayerStartX + layerSpace
predY = actlayerStartY
pgm.add_node(daft.Node("Pred",
                       r"$Prediction$",
                       predX,
                       predY + 0 * outnodeSpace,
                       fixed = True))

# connect input nodes to hidden layer 1
pgm.add_edge("C1", "H1")
pgm.add_edge("C2", "H1")
pgm.add_edge("C3", "H1")
pgm.add_edge("C4", "H1")
pgm.add_edge("C5", "H1")
pgm.add_edge("C6", "H1")
pgm.add_edge("C7", "H1")
pgm.add_edge("C8", "H1")

pgm.add_edge("C1", "H2")
pgm.add_edge("C2", "H2")
pgm.add_edge("C3", "H2")
pgm.add_edge("C4", "H2")
pgm.add_edge("C5", "H2")
pgm.add_edge("C6", "H2")
pgm.add_edge("C7", "H2")
pgm.add_edge("C8", "H2")

pgm.add_edge("C1", "H3")
pgm.add_edge("C2", "H3")
pgm.add_edge("C3", "H3")
pgm.add_edge("C4", "H3")
pgm.add_edge("C5", "H3")
pgm.add_edge("C6", "H3")
pgm.add_edge("C7", "H3")
pgm.add_edge("C8", "H3")

pgm.add_edge("C1", "H4")
pgm.add_edge("C2", "H4")
pgm.add_edge("C3", "H4")
pgm.add_edge("C4", "H4")
pgm.add_edge("C5", "H4")
pgm.add_edge("C6", "H4")
pgm.add_edge("C7", "H4")
pgm.add_edge("C8", "H4")

pgm.add_edge("C1", "H5")
pgm.add_edge("C2", "H5")
pgm.add_edge("C3", "H5")
pgm.add_edge("C4", "H5")
pgm.add_edge("C5", "H5")
pgm.add_edge("C6", "H5")
pgm.add_edge("C7", "H5")
pgm.add_edge("C8", "H5")

pgm.add_edge("C1", "H6")
pgm.add_edge("C2", "H6")
pgm.add_edge("C3", "H6")
pgm.add_edge("C4", "H6")
pgm.add_edge("C5", "H6")
pgm.add_edge("C6", "H6")
pgm.add_edge("C7", "H6")
pgm.add_edge("C8", "H6")

pgm.add_edge("C1", "H7")
pgm.add_edge("C2", "H7")
pgm.add_edge("C3", "H7")
pgm.add_edge("C4", "H7")
pgm.add_edge("C5", "H7")
pgm.add_edge("C6", "H7")
pgm.add_edge("C7", "H7")
pgm.add_edge("C8", "H7")

pgm.add_edge("C1", "H8")
pgm.add_edge("C2", "H8")
pgm.add_edge("C3", "H8")
pgm.add_edge("C4", "H8")
pgm.add_edge("C5", "H8")
pgm.add_edge("C6", "H8")
pgm.add_edge("C7", "H8")
pgm.add_edge("C8", "H8")

# connect hidden layer 1 to output nodes
pgm.add_edge("H1", "O1")
pgm.add_edge("H2", "O1")
pgm.add_edge("H3", "O1")
pgm.add_edge("H4", "O1")
pgm.add_edge("H5", "O1")
pgm.add_edge("H6", "O1")
pgm.add_edge("H7", "O1")
pgm.add_edge("H8", "O1")

pgm.add_edge("H1", "O2")
pgm.add_edge("H2", "O2")
pgm.add_edge("H3", "O2")
pgm.add_edge("H4", "O2")
pgm.add_edge("H5", "O2")
pgm.add_edge("H6", "O2")
pgm.add_edge("H7", "O2")
pgm.add_edge("H8", "O2")

pgm.add_edge("H1", "O3")
pgm.add_edge("H2", "O3")
pgm.add_edge("H3", "O3")
pgm.add_edge("H4", "O3")
pgm.add_edge("H5", "O3")
pgm.add_edge("H6", "O3")
pgm.add_edge("H7", "O3")
pgm.add_edge("H8", "O3")

pgm.add_edge("H1", "O4")
pgm.add_edge("H2", "O4")
pgm.add_edge("H3", "O4")
pgm.add_edge("H4", "O4")
pgm.add_edge("H5", "O4")
pgm.add_edge("H6", "O4")
pgm.add_edge("H7", "O4")
pgm.add_edge("H8", "O4")

# connect output nodes to activation function
pgm.add_edge("O1", "Act")
pgm.add_edge("O2", "Act")
pgm.add_edge("O3", "Act")
pgm.add_edge("O4", "Act")

# connect activation function to prediction
pgm.add_edge("Act", "Pred")

pass
```
