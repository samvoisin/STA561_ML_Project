Single hidden layer from paper

```python
pgm = daft.PGM(shape=[5, 5], origin=[0, -0.5])

# input nodes; these are the channels
pgm.add_node(daft.Node("C1", r"$C1$", 0.5, 0))
pgm.add_node(daft.Node("C2", r"$C2$", 0.5, 0.55))
pgm.add_node(daft.Node("C3", r"$C3$", 0.5, 1.10))
pgm.add_node(daft.Node("C4", r"$C4$", 0.5, 1.65))
pgm.add_node(daft.Node("C5", r"$C5$", 0.5, 2.2))
pgm.add_node(daft.Node("C6", r"$C6$", 0.5, 2.75))
pgm.add_node(daft.Node("C7", r"$C7$", 0.5, 3.30))
pgm.add_node(daft.Node("C8", r"$C8$", 0.5, 3.85))

# hidden layer1
pgm.add_node(daft.Node("H1", r"$H_{1}$", 3, 0))
pgm.add_node(daft.Node("H2", r"$H_{2}$", 3, 0.6))
pgm.add_node(daft.Node("n", r"$n$", 4, 0, fixed=True))
pgm.add_node(daft.Node("y", r"$y$", 4, 0, observed=True))

# connect input nodes to hidden layer 1
pgm.add_edge("C1", "H1")
pgm.add_edge("C2", "H1")
pgm.add_edge("C3", "H1")
pgm.add_edge("C4", "H1")
pgm.add_edge("C5", "H1")
pgm.add_edge("C6", "H1")
pgm.add_edge("C7", "H1")
pgm.add_edge("C8", "H1")

pgm.add_edge("C1", "H2")
pgm.add_edge("C2", "H2")
pgm.add_edge("C3", "H2")
pgm.add_edge("C4", "H2")
pgm.add_edge("C5", "H2")
pgm.add_edge("C6", "H2")
pgm.add_edge("C7", "H2")
pgm.add_edge("C8", "H2")

pgm.add_edge("n", "y")
pgm.add_edge("H1", "y")

pgm.render()

pass
```
